```python
def add_deep_methods(data_all, deep_methods, datasets=["PCPNet_vardensity_striped", "PCPNet_vardensity_gradient", "PCPNet_no_noise", "PCPNet_low_noise", "PCPNet_med_noise", "PCPNet_high_noise", "avg"]):
    for method in deep_methods:
        name = method["name"]
        for dataset in datasets:
            dataset_name = dataset
            value = method[dataset]
            if dataset == "avg":
                continue
            new_line = pd.DataFrame({"dataset": [dataset_name], "Method": [name], "normal mean": [value], "Radius":[-1]})
            data_all = pd.concat([data_all, new_line], ignore_index=True)
    return data_all
```

```python
import math

def grab_best_methods(data, radii, properties, methods):
    best_methods = {}
    current_best_method = None
    current_best_radius = None
    current_best_value = math.inf
    for radius in radii+[-1] :
        for ( property, method_list ) in zip(properties, methods) :
            # print ("Radius", radius, "Property", property)
            # print ("Methods", method_list)

            data_current = data[data["Radius"] == radius]
            # mean the result for each method
            data_current = data_current[data_current["Method"].isin(method_list)]
            for method in method_list :
                if not data_current.empty :
                    mean = data_current[data_current["Method"] == method][property].mean()
                    if mean <= current_best_value :
                        current_best_value = mean
                        current_best_method = method
                        current_best_radius = radius
                else:
                    print("No data for method", method)
    best_methods[property] = ( current_best_method, current_best_radius )
    return best_methods
```

```python
def add_reference(method) :
    return "~\\cite{"+tb.references[method][1]+"}"
```

```python
def radii_to_txt(radii) :
    txt = "$["
    for radius in radii :
        txt += str(radius) + ", "
    return txt[:-2] + "]$"
```

```python
# splits need to be as folow : 
# splits = [ ("Name", ["name col1", "name col2", ...] ), ... ]
# splits_dataframe = [("dataframe_name_property", [dataframe_col1, ...]) ...]

# The length of the tab is the number of "name col" in the splits

def grab_data_current (data, method, radius, group, name) :
    data_current = data[(data_all["Method"] == method)]
    data_current = data_current[ data_current["Radius"].isin([radius, -1])]
    data_current = data_current[ data_current[group] == name ]

    return data_current

def generate_big_tab (data, splits_name, splits_dataframe, radii, property, methods, caption) :

    # splits_names and splits_dataframe must have the same length
    if len(splits_name) != len(splits_dataframe) :
        print ("Error : splits_name and splits_dataframe must have the same length")
        return
    
    splitting = ""

    for i, split in enumerate(splits_dataframe) :
        if len(split[1]) != len(splits_name[i][1]) :
            print ("Error : splits_dataframe must have the same length for each split")
            return
        splitting += " "+"r"* len(split[1])  

    length_splits= sum([len(split[1]) for split in splits_name])


    output = "\\begin{table*}[h!] \n \\centering \n { \n \\resizebox{\\textwidth}{!}\n{ \n \\begin{tabular}{l"+splitting+"}\n \\toprule\n"
    output += "\\multirow{2}{*}{Methods} "
    for i, (group, group_names) in enumerate(splits_name) :
        if i != len(splits_name) - 1 :
            if i == 0 :
                output += " & \\multicolumn{"+str(len(group_names))+"}{|c|}{"+group+"} "
            else :
                output += " & \\multicolumn{"+str(len(group_names))+"}{c|}{"+group+"} "
        else:
            output += " & \\multicolumn{"+str(len(group_names))+"}{c}{"+group+"} "
    output += " \\\\ \n"
    for group, group_names in splits_name :
        for name in group_names :
            output += " & " + name + " "
    output += " \\\\ \n"
    output += "\\midrule\n"

    for method in methods :
        if not ( method in(data["Method"].unique()) ) :
            continue
        output += "\\midrule\n"
        # method_name = method.replace("_", "\_")
        # method_name = "\\texttt{" + method_name + "}"
        method_name = mt.references[method][0]
        output += "\\multirow{"+str(len(radii))+"}{*}{" + method_name + add_reference(method) +"}"
        # output += "\\multirow{"+str(len(radii))+"}{*}{" + method_name + "~\\cite{TODO}}"
        for radius in radii :
            for group, group_names in splits_dataframe :
                for name in group_names :
                    data_check = data[data[group] == name]
                    # data_current = data[(data_all["Method"] == method) & (data_all["Radius"] == radius) & (data_all[group] == name)]
                    data_current = grab_data_current(data, method, radius, group, name)
                    best_methods = grab_best_methods(data_check, radii, [property], [methods])
                    mean_error = data_current[property].mean()
                    # if bigger than 100, put as power of e
                    if mean_error > 300 :
                        mean_error = "{:.0e}".format(mean_error)
                    else :
                        mean_error = round(mean_error, 4)
                    if method == best_methods[property][0] and ( radius == best_methods[property][1] or best_methods[property][1] == -1 ) :
                        output += " & \\good{" + str(mean_error) + "}"
                    else :
                        output += " & " + str(mean_error)
            output += " \\\\ \n"

    output += "\\bottomrule\n \\end{tabular}\n } \n } \n \\caption{"+caption+"} \n \\label{tab:TODO} \n \\end{table*}"
    return output
```
